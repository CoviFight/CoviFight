```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv('time_series_covid_19_confirmed.csv') 
print(data.shape)
data.head()
```

```python
cdata=data[data["Country/Region"] == 'China']
```

```python
arrM=[]
for j in range(4,20):
    arr=[[0 for i in range(50)] for j in range(50)]
    for i in range(33):
        lat=int(cdata.iloc[i,2])-10
        long=int(cdata.iloc[i,3])-80
        if cdata.iloc[i,j]!=0:
            arr[lat][long]+=math.log(cdata.iloc[i,j])
    arrM+=[arr]        
```

# Total Cases

```python
plt.figure(figsize = (50,50))
for i in range(15):
    plt.subplot(5,3,i+1)
    df_cm = pd.DataFrame(arrM[i])   
    sns.heatmap(df_cm, cmap="Blues")# font size
```

# New Cases

```python
plt.figure(figsize = (50,50))
for i in range(0,15):
    plt.subplot(5,3,i+1)
    old=[]
    if i==0:
        old=[[0 for i in range(50)]for j in range(50)]
    else:
        old=arrM[i-1]
    current=arrM[i]
    change=[[0 for i in range(50)]for j in range(50)]
    for i in range(50):
        for j in range(50):
            if(old[i][j]!=0):
                change[i][j]=current[i][j]/old[i][j]-1
            else:
                change[i][j]=current[i][j]
    df_cm = pd.DataFrame(change)   
    sns.heatmap(df_cm, cmap="Blues")# font size
```
